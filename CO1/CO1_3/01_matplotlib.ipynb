```python
import matplotlib.pyplot as pl
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
pl.plot(x,y,'r')
pl.xlabel('X axis')
pl.ylabel('Y axis')
pl.title('Graph')
pl.show()
```

```python
x=np.arange(0,10)
y=x**2
```

```python
x
```

```python
y
```

```python
pl.plot(x,y,'r')
pl.xlabel('X axis')
pl.ylabel('Y axis')
pl.title('Graph')
pl.show()
```

```python
#pl.subplot(nrows,ncols,plot_number)
pl.subplot(2,3,1)
pl.plot(x,y,'r-')
pl.subplot(2,3,2)
pl.plot(x,y,'g*-')
pl.subplot(2,3,3)
pl.plot(x,y,'y--')
pl.subplot(2,3,4)
pl.plot(x,y,'--')
pl.subplot(2,3,5)
pl.plot(x,y,'b--')
pl.subplot(2,3,6)
pl.plot(x,y,'b--')

```

```python
#create figure(empty canvas)
fig= pl.figure()
#Ad set of axes to figure 
ax=fig.add_axes([0.1,0.1,0.9,0.9])#left,bottom,width,height()0-1

#plot that set of axis
ax.plot(x,y,'g')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Figure1')
```

```python
#create figure(empty canvas)
fig= pl.figure()
#Ad set of axes to figure 
ax=fig.add_axes([0.1,0.1,0.9,0.9])#left,bottom,width,height()0-1

#plot that set of axis
ax.plot(x,y,'g')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Figure1')

fig1= pl.figure()
ax1=fig.add_axes([0.3,0.3,0.5,0.5])
ax1.plot(x,y,'r')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.set_title('Figure2')
```

```python
#Use similar to plt.figure() exept use tuple unpacking to grab fig and axis

fig, ax1=pl.subplots(nrows=1, ncols=2)

#Now use the axis object to add stuff to plot
for a in ax1:
  a.plot(x,y,'b')
  a.set_xlabel('X-axis')
  a.set_ylabel('Y-axis')
  a.set_title('Figure1')
  
fig.tight_layout()
```

```python
fig, ax=pl.subplots(figsize=(5,5),dpi=100)

ax.plot(x,y,'g')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Figure1')
```

```python
fig, ax=pl.subplots(dpi=80)

ax.plot(x,y,'g')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Figure1')
```

#**Legends**  

```python
fig= pl.figure()
ax=fig.add_axes([0.1,0.1,0.9,0.9])
ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Figure1')
ax.legend(loc=1)
ax.legend(loc=2)
```

#Style line color

```python
fig, ax=pl.subplots()

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Figure1')

ax.plot(x,x**2,'b--')
ax.plot(x,x**3,'g.-')
```

```python
fig, ax=pl.subplots()

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Figure1')

ax.plot(x,x+1,'b')
ax.plot(x,x+2,'g')
ax.plot(x,x+3,'r')
ax.plot(x,x+4,'y')
```

#**Line and Marker styles**

```python
fig, ax=pl.subplots(dpi=120)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Figure1')

ax.plot(x,x+9,'b',linewidth=0.25)
ax.plot(x,x+8,'g',linewidth=2.00)
ax.plot(x,x+7,'r',linewidth=3.00)

ax.plot(x,x+6,'b',lw=4,ls='--')
ax.plot(x,x+5,'g',lw=4,ls='-.')
ax.plot(x,x+4,'r',lw=4,ls=':')

ax.plot(x,x+3,'b',lw=4,ls='--',marker='o',markersize=2)
ax.plot(x,x+2,'g',lw=4,ls='-.',marker='o',markersize=4)
ax.plot(x,x+1,'r',lw=4,ls=':',marker='s',markersize=6,markerfacecolor='g')
```
