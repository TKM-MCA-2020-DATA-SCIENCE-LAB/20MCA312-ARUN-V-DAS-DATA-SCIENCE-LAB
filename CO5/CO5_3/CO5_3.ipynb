```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/Reviews.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Score'].value_counts()
```

```python
df['positively rate']=np.where(df['Score']>=3,1,0)
```

```python
df.head()
```

```python
df['positively rate'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df['Summary'],df['positively rate'],random_state=50)
```

```python
print(x_train)
```

```python
df.head()
```

```python
print(y_train)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect=CountVectorizer(min_df=5,ngram_range=(1,2),).fit(x_train)
```

```python
len(vect.get_feature_names())
```

```python
vect.get_feature_names()
```

```python
x_train_vectorized=vect.transform(x_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_vectorized,y_train)
```

```python
Predictions=model.predict(vect.transform(x_test))
```

```python
print(Predictions)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,Predictions))
```

```python
print(classification_report(y_test,Predictions))
```

```python
from sklearn.metrics import roc_auc_score
print('AUC score is',roc_auc_score(y_test,Predictions))
```
